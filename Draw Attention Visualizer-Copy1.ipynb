```python


# from bokeh.charts import Scatter
# from bokeh.sampledata.autompg import autompg as df
from bokeh.plotting import figure
from bokeh.io import show
# print(df)

p = figure(plot_width=400, plot_height=400, background_fill_color="#111111", x_range=(0, 200), y_range=(200, 0), border_fill_color="#111111")
# add a circle renderer with a size, color, and alpha
dark = "#111111"
light = "#DDDDDD"
p.border_fill_color = dark
p.title.text_color = light
p.xaxis.axis_label_text_color = light
p.yaxis.axis_label_text_color = light
p.xaxis.axis_line_color = light
p.yaxis.axis_line_color = light
p.xaxis.major_label_text_color = light
p.yaxis.major_label_text_color = light
p.xaxis.major_tick_line_color = light
p.xaxis.minor_tick_line_color = light
p.yaxis.major_tick_line_color = light
p.yaxis.minor_tick_line_color = light

mu_x_list = mu_x * 25
mu_y_list = [val for val in mu_y for _ in range(0, 25)]
p.circle(mu_x_list, mu_y_list, size=1, color="orange", alpha=0.5)

# show the results
show(p)

# p = Scatter(df, x='mpg', y='hp', title="HP vs MPG",
#             xlabel="Miles Per Gallon", ylabel="Horsepower")

# show(p)
```
